```python
import pandas as pd 
from ntscraper import Nitter 
```

```python
sc = Nitter()
```

```python
tweets = sc.get_tweets('elonmusk', mode ='user',number = 50000)
```

```python
tweets
```

```python
for tweet in tweets['tweets']:
    print(tweet)
    print('========================')
    
```

```python
final=[]
for tweet in tweets['tweets']:
    data = [tweet['link'] , tweet['text'] , tweet['stats']['comments'], tweet['stats']['retweets'], tweet['stats']['quotes'],tweet['stats'][ 'likes'] , tweet['user']['avatar']]
    final.append(data)
```

```python
df = pd.DataFrame(final , columns = ['link', 'text' ,'comments','retweets','quotes','likes' ,'avatar'])
```
